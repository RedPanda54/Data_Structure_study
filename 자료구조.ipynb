**양방향 연결리스트**

```python
# 양방향 연결리스트
class Node:
    def __init__(self, key=None):
        self.key = key
        self.prev = self
        self.next = self
    def __str__(self):
        return str(self.key)

class DoublyLinkedList:
    def __init__(self):
        self.head = Node()

    def __iter__(self):
        v = self.head.next
        while v != self.head:
            yield v
            v = v.next

    def __str__(self):
        return " -> ".join(str(v.key) for v in self)

    def printList(self):
        v = self.head.next
        print("h -> ", end="")
        while v != self.head:
            print(str(v.key)+" -> ", end="")
            v = v.next
        print("h")

    def splice(self, a, b, x):
        if a == None or b == None or x == None:	# dummy node만 있을 경우 return
            return
        ap = a.prev	 # 잘라내야할 노드의 이전 노드를 변수에 저장
        bn = b.next	 # 잘라내야할 노드의 다음 노드를 변수에 저장
        xn = x.next	 # [a..b]를 붙이고 난 다음 노드
        #CUT[a..b]
        ap.next = bn # ap는 bn으로 우향 접근
        bn.prev = ap # bn은 ap로 좌향 접근
        #PASTE [a..b]
        xn.prev = b	 # xn에서 [a..b]로 좌향 접근
        b.next = xn	 # [a..b]에서 xn로 우향 접근
        a.prev = x 	 # [a..b]에서 x로 좌향 접근
        x.next = a	 # x에서 [a..b]로 우향 접근
    
    # 이동
    def moveAfter(self, a, x):	# a를 x의 오른족으로 이동
        self.splice(a, a, x)
    def moveBefore(self, a, x): # a를 x의 왼쪽으로 이동
        self.splice(a, a, x.prev)

    # 삽입
    def insertAfter(self, x, key):	# key값을 가진 노드를 x의 오른쪽에 삽입
        self.moveAfter(Node(key), x)
    def insertBefore(self, x, key): # key값을 가진 노드를 x의 왼쪽에 삽입
        self.moveBefore(Node(key), x)
    def pushFront(self, key):		# dummyhead 다음에 key값을 가진 노드를 삽입
        self.insertAfter(self.head, key)
    def pushBack(self, key):		# dummyhead 이전에 key값을 가진 노드를 삽입
        self.insertBefore(self.head, key)

    # 삭제
    def deleteNode(self, x): # x 노드를 찾아 삭제
        if x == None or x == self.head:return # x가 dummy node일 경우 return
        xp = x.prev			# x 노드의 이전 노드를 변수에 저장
        xn = x.next			# x 노드의 다음 노드를 변수에 저장
        xp.next = x.next	# xp은 x.next로 우향 접근
        xn.prev = x.prev	# xn은 x.prev로 좌향 접근
        del x               # x는 삭제
    def popFront(self):     # dummy node 다음 노드를 삭제
        if self.isEmpty():	# 연결리스트가 비어있다면 (dummy node만 있다면) None을 return 
            return None
        key = self.head.next.key	    # dummy node.next.key
        self.deleteNode(self.head.next) # deleteNode 함수를 호출해서 삭제
        return key	# 삭제한 key값을 return
    def popBack(self):  	# dummy node 이전 노드를 삭제
        if self.isEmpty():	# 연결리스트가 비어있다면 (dummy node만 있다면) None을 return
            return None
        key = self.head.prev.key	    # dummy node.prev.key
        self.deleteNode(self.head.prev)	# deleteNode 함수를 호출해서 삭제
        return key	# 삭제한 key값을 return

    # 탐색
    def search(self, key):	# key값이 들어있는 노드를 탐색
        v = self.head.next
        while v != self.head:	# v가 dummy node.next부터 시작해서 dummy node가 될 때까지
            if v.key == key:	# 그 사이에 찾는 key값의 노드가 있다면 그 노드를 return
                return v
            v = v.next
        return None	            # 없으면 None을 return
    def first(self):		# dummy node.next
        return self.head.next
    def last(self):			# dummy node.prev
        return self.head.prev
    def isEmpty(self):  	# 비어있는지 아닌지
        return self.head.next == self.head

    # 최댓값
    def findMax(self):	    # 연결리스트 안에서 최댓값 찾기
        if self.isEmpty():	# 연결리스트가 비어있다면 (dummy node만 있다면) None을 return
            return
        x = self.head.next
        Max = x.key	            # 스타트 지점의 key값을 변수 Max에 저장
        while x != self.head:	# x노드는 dummy node.next에서 dummy node까지 최댓값을 탐색
            if Max <= x.key:	# Max보다 큰 값이 나타나면 Max를 그 값으로 갱신
                Max = x.key
            x = x.next
        return Max

    # 최댓값 삭제
    def deleteMax(self):	# 최댓값을 찾아서 삭제
        if self.isEmpty():	# 연결리스트가 비어있다면 (dummy node만 있다면) None을 return
            return None
        x = self.head.next
        v = x
        Max = x.key     	    # 스타트 지점의 key값을 변수 Max에 저장
        while x != self.head:	# x노드는 dummy node.next에서 dummy node까지 최댓값을 탐색
            if Max <= x.key:	# Max보다 큰 값이 나타나면 Max를 그 값으로 갱신
                Max = x.key
                v = x	        # Max가 갱신되는 지점을 v노드로 지정
            x = x.next
        self.deleteNode(v)	    # 지정해둔 v노드를 deleteNode를 호출해서 삭제
        return v.key
    
    # 정렬
    def sort(self):	# 오름차순 정렬
        x = DoublyLinkedList()	    # x = 새로운 양방향 연결 리스트
        while not self.isEmpty():	# 기존의 연결 리스트가 빌 때까지 최댓값을 x에 pushFront
            x.pushFront(self.deleteMax())
        return x	# 큰 값부터 앞에서 pushF해준 x는 왼쪽부터 오른쪽까지 오름차순 정렬한 상태
L = DoublyLinkedList()
while True:
    cmd = input().split()
    if cmd[0] == 'pushF':
        L.pushFront(int(cmd[1]))
        print("+ {0} is pushed at Front.".format(cmd[1]))
    elif cmd[0] == 'pushB':
        L.pushBack(int(cmd[1]))
        print("+ {0} is pushed at Back.".format(cmd[1]))
    elif cmd[0] == 'popF':
        key = L.popFront()
        if key == None:
            print("* list is empty")
        else:
            print("- {0} is popped from Front.".format(key))
    elif cmd[0] == 'popB':
        key = L.popBack()
        if key == None:
            print("* list is empty")
        else:
            print("- {0} is popped from Back".format(key))
    elif cmd[0] == 'search':
        v = L.search(int(cmd[1]))
        if v == None: print("* {0} is not foundn so nothing happens".format(cmd[1]))
        else: print("* {0} is found!".format(cmd[1]))
    elif cmd[0] == 'insertA':
        x = L.search(int(cmd[1]))
        if x == None: print("* target node of key {0} doesn't exit".format(cmd[1]))
        else:
            L.insertBefore(x, int(cmd[2]))
            print("+ {0} is inserted Before {1}".format(cmd[2], cmd[1]))
    elif cmd[0] == 'insertB':
        x = L.search(int(cmd[1]))
        if x == None: print("* target node of key {0} doesn't exit".format(cmd[1]))
        else:
            L.insertBefore(x, int(cmd[2]))
            print("+ {0} is inserted Befor {1}".format(cmd[2], cmd[1]))
    elif cmd[0] == 'delete':
        x = L.search(int(cmd[1]))
        if x == None:
            print("- {0} is not found, so nothing happens".format(cmd[1]))
        else:
            L.deleteNode(x)
            print("- {0} is deleted".format(cmd[1]))
    elif cmd[0] == "first":
        print("* {0} is the value at the front".format(L.first()))
    elif cmd[0] == "last":
        print("* {0} is the value at the back".format(L.last()))
    elif cmd[0] == "findMax":
        m = L.findMax()
        if m == None:
            print("Empty list!")
        else:
            print("Max key is", m)
    elif cmd[0] == "deleteMax":
        m = L.deleteMax()
        if m == None:
            print("Empty list!")
        else:
            print("Max key", m, "is deleted.")
    elif cmd[0] == 'sort':
        L = L.sort()
        L.printList()
    elif cmd[0] == 'print':
        L.printList()
    elif cmd[0] == 'exit':
        break
    else:
        print("* not allowed command. enter a proper command!")
```

```python
# 대표 정답 (조교님.ver)
class Node:
    def __init__(self, key=None):
        self.key = key
        self.prev = self
        self.next = self
    def __str__(self):
        return str(self.key)

class DoublyLinkedList:
    def __init__(self):
        self.head = Node()

    def __iter__(self):
        v = self.head.next
        while v != self.head:
            yield v
            v = v.next

    def __str__(self):
        return " -> ".join(str(v.key) for v in self)

    def printList(self):
        v = self.head.next
        print("h -> ", end="")
        while v != self.head:
            print(str(v.key)+" -> ", end="")
            v = v.next
        print("h")
	def splice(self, a, b, x):
		if a == None or b == None or x == None:
			return
		
		a.prev.next = b.next
		b.next.prev = a.prev
		
		b.next = x.next
		x.next.prev = b
		x.next = a
		a.prev = x
		
	def moveAfter(self, a, x):
		if a == None or x == None:
			return
		self.splice(a, a, x)
	
	def moveBefore(self, a, x):
		if a == None or x == None:
			return
		self.splice(a, a, x.prev)
	
	def insertAfter(self, a, key):
		if a == None:
			return
		node = Node(key)
		self.splice(node, node, a)
		
	def insertBefore(self, a, key):
		if a == None:
			return
		node = Node(key)
		self.splice(node, node, a.prev)
		
	def pushFront(self, key):
		self.insertAfter(self.head, key)
	
	def pushBack(self, key):
		self.insertBefore(self.head, key)
		
	def search(self, key):
		if self.isEmpty():
			return
		n = self.head.next
		while n !=self.head:
			if n.key == key:
				return n
			n = n.next
		return 
		
	def isEmpty(self):
		return self.head.next == self.head
	
	def first(self):
		return self.head.next
	
	def last(self):
		return self.head.prev
	
	def deleteNode(self, x):
		if x == None or self.isEmpty():
			return
		x.prev.next = x.next
		x.next.prev = x.prev
	
	def popFront(self):
		if self.isEmpty():
			return
		key = self.head.next.key
		self.deleteNode(self.head.next)
		return key
	
	def popBack(self):
		if self.isEmpty():
			return
		key = self.head.prev.key
		self.deleteNode(self.head.prev)
		return key
	
	def printList(self):
		ch = 'h -> '
		n = self.head.next
		while n != self.head:
			ch += f'{n.key} -> '
			n = n.next
		ch += 'h'
		print(ch)
	
	def findMax(self):
		if self.isEmpty():
			return
		n = self.head.next
		max_v = n.key
		while n != self.head:
			if n.key > max_v:
				max_v = n.key
			n = n.next
		return max_v
	
	def deleteMax(self):
		if self.isEmpty():
			return
		max_v = self.findMax()
		self.deleteNode(self.search(max_v))
		return max_v

	def sort(self):
		new_dl = DoublyLinkedList()
		while self.isEmpty() == False:
			new_dl.pushFront(self.deleteMax())
		return new_dl
```

**가장 긴 균형 부문자열 찾기**

```python
def solve(A):
	# code here
	length = 0 # 부문자열 길이가 되어줄 변수
	count_diff = 0 # 0과 1의 개수의 차이
	count_dict = {0: -1} # 0과 1의 개수를 저장할 딕셔너리
	
	for i in range(len(A)):
		if A[i] == '0': # 0이라면 +1
			count_diff += 1
		else: # 1이라면 -1
			count_diff -= 1
			
		if count_diff not in count_dict: # 딕셔너리에 0과 1의 차이가 나온 인덱스를 저장
			count_dict[count_diff] = i
		else:
			# 이미 가지고 있는 index의 차이값과 현재 index의 차이값중 큰 값을 length로 저장
			length = max(length, i - count_dict[count_diff])
	return length
	
A = input().strip()
print(solve(A))

'''
이번 과제에서는 0과 1의 개수를 저장하는데 Dictionary,
그리고 부문자열을 찾는 과정에서 for 반복문을 사용하였다.
알아본 바로는 Dictionary는 O(1)만큼의 수행시간이 걸린다고 한다.
따라서 변수를 저장하는데에 O(1)만큼의 수행시간이 걸렸고,
수업내용을 토대로 for 반복문에서는 O(len(A))만큼의 시간이 걸렸음을 확인할수 있다.
A의 문자열의 개수가 n개라고 한다면
우리가 만든 코드의 총 수행시간은 O(n)으로 결론지을 수 있다.
'''
```

**힙(Heap)**

```python
class Heap:
	def __init__(self, L=[]):
		self.A = L
	def __str__(self):
		return str(self.A)

	def heapify_down(self, k, n):
		while 2*k + 1 < n:
			L, R = 2*k + 1, 2*k + 2
			if self.A[L] > self.A[k]:
				m = L
			else:
				m = k
			if R < n and self.A[R] > self.A[m]:
				m = R
			if m != k:
				self.A[k], self.A[m] = self.A[m], self.A[k]
				k = m
			else:
				break
	
	def make_heap(self):
		n = len(self.A)
		for k in range(n-1, -1, -1):
			self.heapify_down(k, n)
	
	def heap_sort(self):
		n = len(self.A)
		for k in range(len(self.A)-1, -1, -1):
			self.A[0], self.A[k] = self.A[k], self.A[0]
			n = n-1
			self.heapify_down(0, n)
	
S = [int(x) for x in input().split()]
H = Heap(S)
H.make_heap()
H.heap_sort()
print(H)
```

```python
# 변수 초기화 함수
# 각 함수를 호출해 답을 낸뒤 사용한 변수는 다시 초기화시킨다.
# 단, result는 방문시간 결과값만을 저장한 변수이기 때문에 초기화시키지 않는다.
def initialization():
	global current, visit_time, stack, cross
	current = root_node
	visit_time = 0
	stack = []
	stack.append(current)
	cross = []

def preorder(v):
	global current, visit_time, result, stack, cross
	while stack:
		current = stack.pop()
		visit_time += 1
		result = visit_time
		if left[current] != 0 and right[current] != 0:
			cross.append(right[current])
		if current == v:
			initialization()
			return result
		if left[current] != 0:
			stack.append(left[current])
			continue
		elif right[current] != 0:
			stack.append(right[current])
			continue
		else: # left[current] == 0 and right[current] == 0
			current = cross.pop()
			stack.append(current)
	initialization()
	return result

def postorder(v):
	global current, visit_time, result, stack, cross
	n = len(parent)
	while stack:
		current = stack.pop()
		visit_time += 1
		result = visit_time
		if left[current] != 0 and right[current] != 0:
			cross.append(left[current])
		if current == v:
			initialization()
			return (n - result)
		if right[current] != 0:
			stack.append(right[current])
			continue
		elif left[current] != 0:
			stack.append(left[current])
			continue
		else: # left[current] == 0 and right[current]
			current = cross.pop()
			stack.append(current)
	initialization()
	return (n - result)
	
# 노드의 depth 계산 함수
def depth(v):
	level = 0	# 노드의 level
	while parent[v] != 0: # v노드에서 시작해서 루트 노드까지
		if v == root_node:	# v가 루트 노드일 경우
			break	# 초기에 미리 break를 걸어서 level이 더해지지 못하게 방지
		level += 1
		v = parent[v]
	return level

# u는 v의 조상 노드인가?
def is_ancestor(u, v):
	if depth(v) <= depth(u):
		return False
	while parent[v] != 0: # v노드에서 시작해서 루트 노드까지
		if u == parent[v]:	# 도중에 u가 있으면 True
			return True
		v = parent[v]
	return False	# 루트 노드까지 돌았는데 u가 없으면 False

# LCA
def lca(u, v):
	depth_u = depth(u)
	depth_v = depth(v)
	
	if depth_v > depth_u:
		u, v = v, u
		depth_u, depth_v = depth_v, depth_u
		
	while depth_u != depth_v:
		u = parent[u]
		depth_u -= 1
		
	while u != v:
		u = parent[u]
		v = parent[v]
	
	return u

# 입력 부분
n = int(input())	# 노드의 개수

parent = [0] * (n+1)	# 부모 리스트
left = [0] * (n+1)	  # 왼쪽 자식 리스트
right = [0] * (n+1)	  # 오른쪽 자식 리스트

for _ in range(n):
	v, l, r = map(int, input().split())	# 노드 번호, 왼쪽 자식 번호, 오른쪽 자식 번호 입력
	parent[l] = v	# 노드 v는 왼쪽 자식의 부모
	parent[r] = v	# 노드 v는 오른쪽 자식의 부모
	left[v] = l	  # 노드 v의 왼쪽 자식
	right[v] = r	# 노드 v의 오른쪽 자식

# 전처리 부분
root_node = 0	          # 루트 노드
for i in range(1, n+1):
	if parent[i] == 0:	  # 부모가 없는 노드가 루트 노드
		root_node = i
		break

current = root_node # 현재 위치를 지정. 시작은 루트 노드
visit_time = 0	# 방문시간
result = 0	    # 오직 방문시간 결과만을 위한 변수
stack = []	    # preorder와 postorder를 위한 스택
stack.append(current)	# stack안에 root_node를 가지고 시작
cross = []	    # 갈림길에서 사용하는 스택(crossroad)

# 
# 명령 처리 부분으로 아래는 수정 하지 말 것!
#
while True:
	cmd = input().split()
	if cmd[0] == 'exit':
		break
	elif cmd[0] == 'preorder':
		res = preorder(int(cmd[1]))
		print(f"  > preorder({int(cmd[1])}) = {res}")
	elif cmd[0] == 'postorder':
		res = postorder(int(cmd[1]))
		print(f"  > postorder({int(cmd[1])}) = {res}")
	elif cmd[0] == 'depth':
		res = depth(int(cmd[1]))
		print(f"  > depth({int(cmd[1])}) = {res}")
	elif cmd[0] == 'is_ancestor':
		res = is_ancestor(int(cmd[1]), int(cmd[2]))
		print(f"  > {int(cmd[1])} is {'an' if res else 'not an'} ancestor of {int(cmd[2])}")
	elif cmd[0] == 'lca':
		res = lca(int(cmd[1]), int(cmd[2]))
		print(f"  > lca({int(cmd[1])}, {int(cmd[2])}) = {res}")
	else:
		print("illegal command")
```

**이진탐색트리**

```python
# Node class
class Node:
	def __init__(self, key): # 노드 구성
		self.key = key
		self.parent = self.left = self.right = None
		self.height = 0

	def __str__(self):
		return str(self.key)

# BST class
class BST:
	def __init__(self):
		self.root = None
		self.size = 0
	
	def __len__(self):
		return self.size

	# preorder
	def preorder(self, v):
		if v != None: # M
			print(v.key, end=' ')
			self.preorder(v.left)  # L
			self.preorder(v.right) # R

	# inorder
	def inorder(self, v):
		if v != None: # M
			self.inorder(v.left)  # L
			print(v.key, end=' ')
			self.inorder(v.right) # R
	
	# postorder
	def postorder(self, v):
		if v != None: # M
			self.postorder(v.left)  # L
			self.postorder(v.right) # R
			print(v.key, end=' ')

	# Find Location in tree
	def find_loc(self, key):
		# when tree is empty
		if self.size == 0:
			return None
		
		p = None      # p is parent of v
		v = self.root
		
		while v:
			# key is in tree
			if v.key == key:
				return v
			
			# key is not in tree
			else:
				# v is R
				if v.key < key:
					p = v
					v = v.right
				
				# v is L
				else:
					p = v
					v = v.left
		
		return p

	# Search (탐색)
	def search(self, key):
		# Find Location of parent(key)
		p = self.find_loc(key)
		
		# key is in tree
		if p and p.key == key:
			return p
		
		# key is not in tree
		else:
			return None

	# Insert (삽입)
	def insert(self, key):
		p = self.find_loc(key)
		
		if p == None or p.key != key:
			v = Node(key)
			
			# v is root (= p is None)
			if p == None:
				self.root = v
			
			# v is in tree
			else: # p != None and p.key != key
				v.parent = p
				if p.key >= key:
					p.left = v
				else: # p.key < key
					p.right = v
			self.size += 1
			self.updateHeight(v)
			return v
		else:
			return p # return None

	# For Height
	def updateHeight(self, v):
		# v is not in tree
		if v == None:
			return
		
		# v is in tree
		else:
			v.height = self.getHeight(v) # update v.height
			self.updateHeight(v.parent)  # update v.parent.height
	
	def getHeight(self, v):
		# v is not in tree
		if v == None:
			return -1
		
		# v is in tree
		else:
			leftHeight = self.getHeight(v.left) + 1
			rightHeight = self.getHeight(v.right) + 1
			return max(leftHeight, rightHeight)
	
	# Delete (삭제)
	def deleteByMerging(self, x):
		# x is not in tree
		if x == None:
			return
		
		# x is in tree
		else:
			L, R, pt = x.left, x.right, x.parent
			# L is None
			if L == None:
				c = R
				s = pt # 균형이 깨질 가능성이 있는 첫 번째 노드
			
			# L is not None
			else:
				c = m = L # top node in L
				while m.right: # m is maximum of L
					m = m.right
				m.right = R # R is right child of m
				if R:
					R.parent = m
				s = m
			
			# x is root node
			if self.root == x:
				if c:
					c.parent = None
				self.root = c
			
			# x is not root node
			else:
				if pt.left == x:
					pt.left = c
				elif pt.right == x:
					pt.right = c
				
				if c:
					c.parent = pt
				
			self.size -= 1
			self.updateHeight(s)
			return s # first node that would be rebalanced

	def deleteByCopying(self, x):
		# x is not in tree
		if x == None:
			return
		
		# x is in tree
		else:
			L, R, pt = x.left, x.right, x.parent
			if L == None and R == None:
				# x is root node
				if x == self.root:
					self.root = None
				# x is not root node
				else:
					if pt.left == x:
						pt.left = None
					elif pt.right == x:
						pt.right = None
				self.updateHeight(pt)
				del x
				return pt
			
			# L is not empty
			elif L:
				y = self.pred(x) # y is maximum in L
				x.key = y.key
				
				if y.left:
					y.left.parent = y.parent
				
				# y is top node in L
				if y.parent.left == y:
					y.parent.left = y.left
				
				# y is not top node
				else:
					y.parent.right = y.left
				p = y.parent
				
				self.updateHeight(p)
				del y
				return p
			
			# L is empty
			elif R:
				y = self.succ(x) # y is minimum in R
				x.key = y.key
				
				if y.right:
					y.right.parent = y.parent
				
				# y is top node in R
				if y.parent.left == y:
					y.parent.left = y.right
				# y is not top node
				else:
					y.parent.right = y.right
				p = y.parent
				
				self.updateHeight(p)
				del y
				return p

	# Height
	def height(self, x): # x의 height를 return
		if x == None:
			return -1
		else:
			return x.height

	# Successor
	def succ(self, x):
		if x == None:
			return
		
		# R is not None
		if x.right != None:
			p = x.right
			while p.left:
				p = p.left
			return p
		
		# R is None
		else:
			p = x.parent
			while p and x == p.right:
				x = p
				p = p.parent
			return p # 이미 최댓값인 경우 루트노드까지 올라가서 None이 되는 운명

	# Predecessor
	def pred(self, x):
		if x == None:
			return
		
		# L is not None
		if x.left != None:
			p = x.left
			while p.right:
				p = p.right
			return p
		
		# L is None
		else:
			p = x.parent
			while p and x == p.left:
				x = p
				p = p.parent
			return p # 이미 최솟값인 경우 루트노드까지 올라가서 None이 되는 운명

	# Rotate
	def rotateLeft(self, z):
		if z == None:
			return
		x = z.right
		if x == None:
			return
		b = x.left
		x.parent = z.parent
		if z.parent:
			if z.parent.right == z:
				z.parent.right = x
			else:
				z.parent.left = x
		x.left = z
		z.parent = x
		z.right = b
		if b: b.parent = z
		if z == self.root:
			self.root = x
		self.updateHeight(z)

	def rotateRight(self, z):
		if z == None:
			return
		x = z.left
		if x == None:
			return
		b = x.right
		x.parent = z.parent
		if z.parent:
			if z.parent.left == z:
				z.parent.left = x
			else:
				z.parent.right = x
		x.right = z
		z.parent = x
		z.left = b
		if b: b.parent = z
		if z == self.root:
			self.root = x
		self.updateHeight(z)
	
T = BST()
while True:
	cmd = input().split()
	if cmd[0] == 'insert':
		v = T.insert(int(cmd[1]))
		print("+ {0} is inserted".format(v.key))
	elif cmd[0] == 'deleteC':
		v = T.search(int(cmd[1]))
		T.deleteByCopying(v)
		print("- {0} is deleted by copying".format(int(cmd[1])))
	elif cmd[0] == 'deleteM':
		v = T.search(int(cmd[1]))
		T.deleteByMerging(v)
		print("- {0} is deleted by merging".format(int(cmd[1])))
	elif cmd[0] == 'search':
		v = T.search(int(cmd[1]))
		if v == None:
			print("* {0} is not found!".format(cmd[1]))
		else:
			print("* {0} is found!".format(cmd[1]))
	elif cmd[0] == 'height':
		h = T.height(T.search(int(cmd[1])))
		if h == -1:
			print("= {0} is not found!".format(cmd[1]))
		else:
			print("= {0} has height of {1}".format(cmd[1], h))
	elif cmd[0] == 'succ':
		v = T.succ(T.search(int(cmd[1])))
		if v == None:
			print("> {0} is not found or has no successor".format(cmd[1]))
		else:
			print("> {0}'s successor is {1}".format(cmd[1], v.key))
	elif cmd[0] == 'pred':
		v = T.pred(T.search(int(cmd[1])))
		if v == None:
			print("< {0} is not found or has no predecssor".format(cmd[1]))
		else:
			print("< {0}'s predecssor is {1}".format(cmd[1], v.key))
	elif cmd[0] == 'Rleft':
		v = T.search(int(cmd[1]))
		if v == None:
			print("@ {0} is not found!".format(cmd[1]))
		else:
			T.rotateLeft(v)
			print("@ Rotated left at node {0}".format(cmd[1]))
	elif cmd[0] == 'Rright':
		v = T.search(int(cmd[1]))
		if v == None:
			print("@ {0} is not found!".format(cmd[1]))
		else:
			T.rotateRight(v)
			print("@ Rotated right at node {0}".format(cmd[1]))
	elif cmd[0] == 'preorder':
		T.preorder(T.root)
		print()
	elif cmd[0] == 'postorder':
		T.postorder(T.root)
		print()
	elif cmd[0] == 'inorder':
		T.inorder(T.root)
		print()
	elif cmd[0] == 'exit':
		break
	else:
		print("* not allowed command. enter a proper command!")
```

```python
# Node class
class Node:
	def __init__(self, key): # 노드 구성
		self.key = key
		self.parent = self.left = self.right = None
		self.height = 0

	def __str__(self):
		return str(self.key)

# BST class
class BST:
	def __init__(self):
		self.root = None
		self.size = 0
	
	def __len__(self):
		return self.size

	# preorder
	def preorder(self, v):
		if v != None: # M
			print(v.key, end=' ')
			self.preorder(v.left)  # L
			self.preorder(v.right) # R

	# inorder
	def inorder(self, v):
		if v != None: # M
			self.inorder(v.left)  # L
			print(v.key, end=' ')
			self.inorder(v.right) # R
	
	# postorder
	def postorder(self, v):
		if v != None: # M
			self.postorder(v.left)  # L
			self.postorder(v.right) # R
			print(v.key, end=' ')

	# Find Location in tree
	def find_loc(self, key):
		# when tree is empty
		if self.size == 0:
			return None
		
		p = None      # p is parent of v
		v = self.root
		
		while v:
			# key is in tree
			if v.key == key:
				return v
			
			# key is not in tree
			else:
				# v is R
				if v.key < key:
					p = v
					v = v.right
				
				# v is L
				else:
					p = v
					v = v.left
		
		return p

	# Search (탐색)
	def search(self, key):
		# Find Location of parent(key)
		p = self.find_loc(key)
		
		# key is in tree
		if p and p.key == key:
			return p
		
		# key is not in tree
		else:
			return None

	# Insert (삽입)
	def insert(self, key):
		p = self.find_loc(key)
		
		if p == None or p.key != key:
			v = Node(key)
			
			# v is root (= p is None)
			if p == None:
				self.root = v
			
			# v is in tree
			else: # p != None and p.key != key
				v.parent = p
				if p.key >= key:
					p.left = v
				else: # p.key < key
					p.right = v
			self.size += 1
			self.updateHeight(v)
			return v
		else:
			return p # return None

	# For Height
	def updateHeight(self, v):
		# v is not in tree
		if v == None:
			return
		
		# v is in tree
		else:
			v.height = self.getHeight(v) # update v.height
			self.updateHeight(v.parent)  # update v.parent.height
	
	def getHeight(self, v):
		# v is not in tree
		if v == None:
			return -1
		
		# v is in tree
		else:
			leftHeight = self.getHeight(v.left) + 1
			rightHeight = self.getHeight(v.right) + 1
			return max(leftHeight, rightHeight)
	
	# Delete (삭제)
	def deleteByMerging(self, x):
		# x is not in tree
		if x == None:
			return
		
		# x is in tree
		else:
			L, R, pt = x.left, x.right, x.parent
			# L is None
			if L == None:
				c = R
				s = pt
			
			# L is not None
			else:
				c = m = L # top node in L
				while m.right: # m is maximum of L
					m = m.right
				m.right = R # R is right child of m
				if R:
					R.parent = m
				s = m
			
			# x is root node
			if self.root == x:
				if c:
					c.parent = None
				self.root = c
			
			# x is not root node
			else:
				if pt.left == x:
					pt.left = c
				elif pt.right == x:
					pt.right = c
				
				if c:
					c.parent = pt
				
			self.size -= 1
			self.updateHeight(s)
			return s # first node that would be rebalanced

	def deleteByCopying(self, x):
		# x is not in tree
		if x == None:
			return
		
		# x is in tree
		else:
			L, R, pt = x.left, x.right, x.parent
			if L == None and R == None:
				# x is root node
				if x == self.root:
					self.root = None
				# x is not root node
				else:
					if pt.left == x:
						pt.left = None
					elif pt.right == x:
						pt.right = None
				self.updateHeight(pt)
				del x
				return pt
			
			# L is not empty
			elif L:
				y = self.pred(x) # y is maximum in L
				x.key = y.key
				
				if y.left:
					y.left.parent = y.parent
				
				# y is top node in L
				if y.parent.left == y:
					y.parent.left = y.left
				
				# y is not top node
				else:
					y.parent.right = y.left
				p = y.parent
				
				self.updateHeight(p)
				del y
				return p
			
			# L is empty
			elif R:
				y = self.succ(x) # y is minimum in R
				x.key = y.key
				
				if y.right:
					y.right.parent = y.parent
				
				# y is top node in R
				if y.parent.left == y:
					y.parent.left = y.right
				# y is not top node
				else:
					y.parent.right = y.right
				p = y.parent
				
				self.updateHeight(p)
				del y
				return p

	# Height
	def height(self, x): # x의 height를 return
		if x == None:
			return -1
		else:
			return x.height

	# Successor
	def succ(self, x):
		if x == None:
			return
		
		# R is not None
		if x.right != None:
			p = x.right
			while p.left:
				p = p.left
			return p
		
		# R is None
		else:
			p = x.parent
			while p and x == p.right:
				x = p
				p = p.parent
			return p # 이미 최댓값인 경우 루트노드까지 올라가서 None이 되는 운명

	# Predecessor
	def pred(self, x):
		if x == None:
			return
		
		# L is not None
		if x.left != None:
			p = x.left
			while p.right:
				p = p.right
			return p
		
		# L is None
		else:
			p = x.parent
			while p and x == p.left:
				x = p
				p = p.parent
			return p # 이미 최솟값인 경우 루트노드까지 올라가서 None이 되는 운명

	# Rotate
	def rotateLeft(self, z):
		if z == None:
			return
		x = z.right
		if x == None:
			return
		b = x.left
		x.parent = z.parent
		if z.parent:
			if z.parent.right == z:
				z.parent.right = x
			else:
				z.parent.left = x
		x.left = z
		z.parent = x
		z.right = b
		if b: b.parent = z
		if z == self.root:
			self.root = x
		self.updateHeight(z)

	def rotateRight(self, z):
		if z == None:
			return
		x = z.left
		if x == None:
			return
		b = x.right
		x.parent = z.parent
		if z.parent:
			if z.parent.left == z:
				z.parent.left = x
			else:
				z.parent.right = x
		x.right = z
		z.parent = x
		z.left = b
		if b: b.parent = z
		if z == self.root:
			self.root = x
		self.updateHeight(z)

# AVL class (child class of BST)
class AVL(BST):
	def __init__(self):
		self.root = None
		self.size = 0
		
	def rebalance(self, x, y, z):
		# rotate 1회
		if z.left == y and y.left == x:     # 왼쪽 방향으로 일직선
			self.rotateRight(z)
			return y
		elif z.right == y and y.right == x: # 오른쪽 방향으로 일직선
			self.rotateLeft(z)
			return y
		
		# rotate 2회
		elif z.left == y and y.right == x:  # 삼각형 경우 1
			self.rotateLeft(y)
			self.rotateRight(z)
			return x
		elif z.right == y and y.left == x:  # 삼각형 경우 2
			self.rotateRight(y)
			self.rotateLeft(z)
			return x
	
	# Insert (AVL 삽입)
	def insert(self, key):
		# BST insert
		v = super(AVL, self).insert(key)
		
		# find x, y, z
		if self.size < 3: # self.size == 1 or self.size == 2
			return v
		else: # self.size >= 3
			z = v
			y = x = None
			while z != self.root: # check from v to root
				if abs(self.height(z.left) - self.height(z.right)) > 1:
					break
				x = y
				y = z
				z = z.parent
			
			# z is blanced
			if abs(self.height(z.left) - self.height(z.right)) <= 1:
				return v
			
			# z is unbalanced
			else:
				w = self.rebalance(x, y, z)
				if w.parent == None:
					self.root = w
				return v
	
	# Delete (AVL 삭제)
	def delete(self, u):
		# BST deleteByCopying
		s = self.deleteByCopying(u)
		
		if self.size < 3: # self.size == 1 or self.size == 2
			return
		
		else: # self.size >= 3
			w = s
			while w != None:
				# when w is unbalanced
				if abs(self.height(w.left) - self.height(w.right)) > 1:
					
					# find z
					z = w
					
					# find y
					if self.height(z.left) >= self.height(z.right):
						y = z.left
					else: # self.height(z.left) < self.height(z.right)
						y = z.right
					
					# find x
					if self.height(y.left) >= self.height(y.right):
						x = y.left
					else: # self.height(y.left) < self.height(y.right)
						x = y.right
					
					# rebalancing
					w = self.rebalance(x, y, z)
					w = w.parent
				
				# when w is balanced
				else: # abs(self.height(s.left) - self.height(s.right)) <= 1:
					w = w.parent

T = AVL()
while True:
	cmd = input().split()
	if cmd[0] == 'insert':
		v = T.insert(int(cmd[1]))
		print("+ {0} is inserted".format(v.key))
	elif cmd[0] == 'delete':
		v = T.search(int(cmd[1]))
		T.delete(v)
		print("- {0} is deleted".format(int(cmd[1])))
	elif cmd[0] == 'search':
		v = T.search(int(cmd[1]))
		if v == None:
			print("* {0} is not found!".format(cmd[1]))
		else:
			print("* {0} is found!".format(cmd[1]))
	elif cmd[0] == 'height':
		h = T.height(T.search(int(cmd[1])))
		if h == -1:
			print("= {0} is not found!".format(cmd[1]))
		else:
			print("= {0} has height of {1}".format(cmd[1], h))
	elif cmd[0] == 'succ':
		v = T.succ(T.search(int(cmd[1])))
		if v == None:
			print("> {0} is not found or has no successor".format(cmd[1]))
		else:
			print("> {0}'s successor is {1}".format(cmd[1], v.key))
	elif cmd[0] == 'pred':
		v = T.pred(T.search(int(cmd[1])))
		if v == None:
			print("< {0} is not found or has no predecssor".format(cmd[1]))
		else:
			print("< {0}'s predecssor is {1}".format(cmd[1], v.key))
	elif cmd[0] == 'preorder':
		T.preorder(T.root)
		print()
	elif cmd[0] == 'postorder':
		T.postorder(T.root)
		print()
	elif cmd[0] == 'inorder':
		T.inorder(T.root)
		print()
	elif cmd[0] == 'exit':
		break
	else:
		print("* not allowed command. enter a proper command!")
```
