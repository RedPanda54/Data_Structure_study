```python
# Infix-to-Postfix
class Stack:
    def __init__(self):
        self.items = []
        
    def push(self, val):
        self.items.append(val)
        
    def pop(self):
        try:
            return self.items.pop()
        except IndexError:
            print("Stack is empty")
            
    def top(self):
        try:
            return self.items[-1]
        except IndexError:
            print("Stack is empty")
            
    def __len__(self):
        return len(self.items)
    
    def isEmpty(self):
        return self.__len__() == 0
    
def infix_to_postfix(infix):
    opstack = Stack()
    outstack = []
    token_list = infix.split()
    # 연산자의 우선순위 설정
    prec = {'(':0, ')':0, '+':1, '-':1, '*':2, '/':2, '^':3}
    for token in token_list:
        if token == '(':
            opstack.push(token)
        elif token == ')':
            while opstack.top() != '(':
                outstack.append(opstack.pop())
            opstack.pop()
        elif token in '+-/*^':
            while opstack.isEmpty() != True:
                if prec[opstack.top()] >= prec[token]:
                    outstack.append(opstack.pop())
                else:
                    break
            opstack.push(token)
        else: # operand일 때
            outstack.append(token)
            
    while opstack.isEmpty() != True:
        outstack.append(opstack.pop())
    # opstack에 남은 모든 연산자를 pop 후 outstack에 append
    return " ".join(outstack)

infix_expr = input()
postfix_expr = infix_to_postfix(infix_expr)
print(postfix_expr)
```

```python
class Stack: # Stack 선언
	def __init__(self): # 데이터 저장을 위해 리스트 준비
		self.items = [] # 빈 리스트
		
	def push(self, val): # 삽입
		self.items.append(val)
	
	def pop(self): # 삭제
		try:
			return self.items.pop()
		except IndexError: # pop할 item이 없다면 IndexError 발생
			print("Stack is empty") # IndexError일 시에 출력
			
	def top(self): # 가장 위의 값을 출력
		try:
			return self.items[-1]
		except IndexError:
			print("Stack is empty")
			
	def __len__(self): # stack의 item 수 반환
		return len(self.items)
	
	def isEmpty(self): # 비었는지 아닌지지
		return len(self) == 0
	
# 입력값을 피연산자와 연산자로 구분해서 리스트화
def get_token_list(expr):
	our_expr = list(expr)         # 입력받은 expr를 리스트에 넣는다.
	infix = []                    # return 받을 리스트
	numbers = list('0123456789.') # 피연산자 리스트 (소수점 포함)
	possible = list('+-*/^() ')   # 피연산자를 제외한 문자 리스트 (연산자, 괄호, 공백)
	
	i = 0
	while i < len(our_expr):
		j = 1
		if our_expr[i] in numbers:         # 현재 index가 피연산자라면??
			while i + j < len(our_expr):
				if our_expr[i + j] in numbers: # 다음 index가 피연산자라면 묶는다.
					j += 1
				else:                          # 다음 index가 피연산자가 아니라면 묶지 않는다.
					break
			# 연산자 이전의 피연산자 전체를 묶음 처리해주고 해당 문자를 infix에 넣는다.
			infix.append(''.join(our_expr[i:i + j]))
			i += j
		else:                              # 현재 index가 피연산자가 아니라면??
			if our_expr[i] in possible:      # 쓸 수 있는 문자는 infix에 넣어준다.
				infix.append(our_expr[i])
				i += 1
			else:                            # 쓰지 못하는 문자는 걸러낸다.
				print(our_expr[i], "은(는) 올바르지 못한 기호입니다.")
				i += 1
				exit()
	return infix
	
# infix를 postfix로 바꾸는 함수
def infix_to_postfix(infix):
	opstack = Stack() # 연산자 구분을 위해 Stack 선언
	outstack = []     # postfix가 되어줄 리스트
	
	# 연산자의 우선순위 설정
	prec = {'(':0, ')':0, '+':1, '-':1, '*':2, '/':2, '^':3}
	
	for token in infix:
		if token == '(':
			opstack.push(token)
		elif token == ')':
			while opstack.top() != '(': # '('를 만날 때까지 opstack에서 연산자를 outstack으로 옮긴다.
				outstack.append(opstack.pop())
			opstack.pop()
		elif token in '+-/*^':
			while opstack.isEmpty() != True:
				if prec[opstack.top()] >= prec[token]: # 우선 순위에 맞춰 처리된다.
					outstack.append(opstack.pop())
				else:
					break
			opstack.push(token)
		else: # 피연산자와 공백은 그대로 outstack에 넣는다.
			outstack.append(token)
	# opstack에 남은 모든 연산자는 지우고 차례대로 outstack에 넣는다.
	while opstack.isEmpty() != True:
		outstack.append(opstack.pop())
	return outstack

# postfix 수식 계산 함수
def compute_postfix(postfix):
	process = Stack()
	
	for token in postfix:
		num1 = 0  # 수식에 쓰이는 두 실수를 초기화
		num2 = 0
		if token == ' ':                # 공백이면 무시한다.
			continue
		else:                           # 공백이 아니라면??
			process.push(token)           # Stack에 넣어준다.
			if token == '+':              # '+'의 경우
				process.pop()               # '+'를 지운다.
				num1 = float(process.pop()) # '+'이전의 피연산자 문자를 실수화 시킨다.
				num2 = float(process.pop())
				process.push(num2 + num1)   # 계산한다.
			elif token == '-':            # 반복
				process.pop()
				num1 = float(process.pop())
				num2 = float(process.pop())
				process.push(num2 - num1)
			elif token == '*':
				process.pop()
				num1 = float(process.pop())
				num2 = float(process.pop())
				process.push(num2 * num1)
			elif token == '/':
				process.pop()
				num1 = float(process.pop())
				num2 = float(process.pop())
				process.push(num2 / num1)
			elif token == '^':
				process.pop()
				num1 = float(process.pop())
				num2 = float(process.pop())
				process.push(num2 ** num1)
	return process.pop() # 최종적으로 Stack에 남아있는 정답을 return 한다.
	
expr = input()
value = compute_postfix(infix_to_postfix(get_token_list(expr)))
print(round(value,4))
```

**양방향 연결리스트**

```python
# 양방향 연결리스트
class Node:
    def __init__(self, key=None):
        self.key = key
        self.prev = self
        self.next = self
    def __str__(self):
        return str(self.key)

class DoublyLinkedList:
    def __init__(self):
        self.head = Node()

    def __iter__(self):
        v = self.head.next
        while v != self.head:
            yield v
            v = v.next

    def __str__(self):
        return " -> ".join(str(v.key) for v in self)

    def printList(self):
        v = self.head.next
        print("h -> ", end="")
        while v != self.head:
            print(str(v.key)+" -> ", end="")
            v = v.next
        print("h")

    def splice(self, a, b, x):
        if a == None or b == None or x == None:	# dummy node만 있을 경우 return
            return
        ap = a.prev	 # 잘라내야할 노드의 이전 노드를 변수에 저장
        bn = b.next	 # 잘라내야할 노드의 다음 노드를 변수에 저장
        xn = x.next	 # [a..b]를 붙이고 난 다음 노드
        #CUT[a..b]
        ap.next = bn # ap는 bn으로 우향 접근
        bn.prev = ap # bn은 ap로 좌향 접근
        #PASTE [a..b]
        xn.prev = b	 # xn에서 [a..b]로 좌향 접근
        b.next = xn	 # [a..b]에서 xn로 우향 접근
        a.prev = x 	 # [a..b]에서 x로 좌향 접근
        x.next = a	 # x에서 [a..b]로 우향 접근
    
    # 이동
    def moveAfter(self, a, x):	# a를 x의 오른족으로 이동
        self.splice(a, a, x)
    def moveBefore(self, a, x): # a를 x의 왼쪽으로 이동
        self.splice(a, a, x.prev)

    # 삽입
    def insertAfter(self, x, key):	# key값을 가진 노드를 x의 오른쪽에 삽입
        self.moveAfter(Node(key), x)
    def insertBefore(self, x, key): # key값을 가진 노드를 x의 왼쪽에 삽입
        self.moveBefore(Node(key), x)
    def pushFront(self, key):		# dummyhead 다음에 key값을 가진 노드를 삽입
        self.insertAfter(self.head, key)
    def pushBack(self, key):		# dummyhead 이전에 key값을 가진 노드를 삽입
        self.insertBefore(self.head, key)

    # 삭제
    def deleteNode(self, x): # x 노드를 찾아 삭제
        if x == None or x == self.head:return # x가 dummy node일 경우 return
        xp = x.prev			# x 노드의 이전 노드를 변수에 저장
        xn = x.next			# x 노드의 다음 노드를 변수에 저장
        xp.next = x.next	# xp은 x.next로 우향 접근
        xn.prev = x.prev	# xn은 x.prev로 좌향 접근
        del x               # x는 삭제
    def popFront(self):     # dummy node 다음 노드를 삭제
        if self.isEmpty():	# 연결리스트가 비어있다면 (dummy node만 있다면) None을 return 
            return None
        key = self.head.next.key	    # dummy node.next.key
        self.deleteNode(self.head.next) # deleteNode 함수를 호출해서 삭제
        return key	# 삭제한 key값을 return
    def popBack(self):  	# dummy node 이전 노드를 삭제
        if self.isEmpty():	# 연결리스트가 비어있다면 (dummy node만 있다면) None을 return
            return None
        key = self.head.prev.key	    # dummy node.prev.key
        self.deleteNode(self.head.prev)	# deleteNode 함수를 호출해서 삭제
        return key	# 삭제한 key값을 return

    # 탐색
    def search(self, key):	# key값이 들어있는 노드를 탐색
        v = self.head.next
        while v != self.head:	# v가 dummy node.next부터 시작해서 dummy node가 될 때까지
            if v.key == key:	# 그 사이에 찾는 key값의 노드가 있다면 그 노드를 return
                return v
            v = v.next
        return None	            # 없으면 None을 return
    def first(self):		# dummy node.next
        return self.head.next
    def last(self):			# dummy node.prev
        return self.head.prev
    def isEmpty(self):  	# 비어있는지 아닌지
        return self.head.next == self.head

    # 최댓값
    def findMax(self):	    # 연결리스트 안에서 최댓값 찾기
        if self.isEmpty():	# 연결리스트가 비어있다면 (dummy node만 있다면) None을 return
            return
        x = self.head.next
        Max = x.key	            # 스타트 지점의 key값을 변수 Max에 저장
        while x != self.head:	# x노드는 dummy node.next에서 dummy node까지 최댓값을 탐색
            if Max <= x.key:	# Max보다 큰 값이 나타나면 Max를 그 값으로 갱신
                Max = x.key
            x = x.next
        return Max

    # 최댓값 삭제
    def deleteMax(self):	# 최댓값을 찾아서 삭제
        if self.isEmpty():	# 연결리스트가 비어있다면 (dummy node만 있다면) None을 return
            return None
        x = self.head.next
        v = x
        Max = x.key     	    # 스타트 지점의 key값을 변수 Max에 저장
        while x != self.head:	# x노드는 dummy node.next에서 dummy node까지 최댓값을 탐색
            if Max <= x.key:	# Max보다 큰 값이 나타나면 Max를 그 값으로 갱신
                Max = x.key
                v = x	        # Max가 갱신되는 지점을 v노드로 지정
            x = x.next
        self.deleteNode(v)	    # 지정해둔 v노드를 deleteNode를 호출해서 삭제
        return v.key
    
    # 정렬
    def sort(self):	# 오름차순 정렬
        x = DoublyLinkedList()	    # x = 새로운 양방향 연결 리스트
        while not self.isEmpty():	# 기존의 연결 리스트가 빌 때까지 최댓값을 x에 pushFront
            x.pushFront(self.deleteMax())
        return x	# 큰 값부터 앞에서 pushF해준 x는 왼쪽부터 오른쪽까지 오름차순 정렬한 상태
L = DoublyLinkedList()
while True:
    cmd = input().split()
    if cmd[0] == 'pushF':
        L.pushFront(int(cmd[1]))
        print("+ {0} is pushed at Front.".format(cmd[1]))
    elif cmd[0] == 'pushB':
        L.pushBack(int(cmd[1]))
        print("+ {0} is pushed at Back.".format(cmd[1]))
    elif cmd[0] == 'popF':
        key = L.popFront()
        if key == None:
            print("* list is empty")
        else:
            print("- {0} is popped from Front.".format(key))
    elif cmd[0] == 'popB':
        key = L.popBack()
        if key == None:
            print("* list is empty")
        else:
            print("- {0} is popped from Back".format(key))
    elif cmd[0] == 'search':
        v = L.search(int(cmd[1]))
        if v == None: print("* {0} is not foundn so nothing happens".format(cmd[1]))
        else: print("* {0} is found!".format(cmd[1]))
    elif cmd[0] == 'insertA':
        x = L.search(int(cmd[1]))
        if x == None: print("* target node of key {0} doesn't exit".format(cmd[1]))
        else:
            L.insertBefore(x, int(cmd[2]))
            print("+ {0} is inserted Before {1}".format(cmd[2], cmd[1]))
    elif cmd[0] == 'insertB':
        x = L.search(int(cmd[1]))
        if x == None: print("* target node of key {0} doesn't exit".format(cmd[1]))
        else:
            L.insertBefore(x, int(cmd[2]))
            print("+ {0} is inserted Befor {1}".format(cmd[2], cmd[1]))
    elif cmd[0] == 'delete':
        x = L.search(int(cmd[1]))
        if x == None:
            print("- {0} is not found, so nothing happens".format(cmd[1]))
        else:
            L.deleteNode(x)
            print("- {0} is deleted".format(cmd[1]))
    elif cmd[0] == "first":
        print("* {0} is the value at the front".format(L.first()))
    elif cmd[0] == "last":
        print("* {0} is the value at the back".format(L.last()))
    elif cmd[0] == "findMax":
        m = L.findMax()
        if m == None:
            print("Empty list!")
        else:
            print("Max key is", m)
    elif cmd[0] == "deleteMax":
        m = L.deleteMax()
        if m == None:
            print("Empty list!")
        else:
            print("Max key", m, "is deleted.")
    elif cmd[0] == 'sort':
        L = L.sort()
        L.printList()
    elif cmd[0] == 'print':
        L.printList()
    elif cmd[0] == 'exit':
        break
    else:
        print("* not allowed command. enter a proper command!")
```

```python
# 대표 정답 (조교님.ver)
class Node:
    def __init__(self, key=None):
        self.key = key
        self.prev = self
        self.next = self
    def __str__(self):
        return str(self.key)

class DoublyLinkedList:
    def __init__(self):
        self.head = Node()

    def __iter__(self):
        v = self.head.next
        while v != self.head:
            yield v
            v = v.next

    def __str__(self):
        return " -> ".join(str(v.key) for v in self)

    def printList(self):
        v = self.head.next
        print("h -> ", end="")
        while v != self.head:
            print(str(v.key)+" -> ", end="")
            v = v.next
        print("h")
	def splice(self, a, b, x):
		if a == None or b == None or x == None:
			return
		
		a.prev.next = b.next
		b.next.prev = a.prev
		
		b.next = x.next
		x.next.prev = b
		x.next = a
		a.prev = x
		
	def moveAfter(self, a, x):
		if a == None or x == None:
			return
		self.splice(a, a, x)
	
	def moveBefore(self, a, x):
		if a == None or x == None:
			return
		self.splice(a, a, x.prev)
	
	def insertAfter(self, a, key):
		if a == None:
			return
		node = Node(key)
		self.splice(node, node, a)
		
	def insertBefore(self, a, key):
		if a == None:
			return
		node = Node(key)
		self.splice(node, node, a.prev)
		
	def pushFront(self, key):
		self.insertAfter(self.head, key)
	
	def pushBack(self, key):
		self.insertBefore(self.head, key)
		
	def search(self, key):
		if self.isEmpty():
			return
		n = self.head.next
		while n !=self.head:
			if n.key == key:
				return n
			n = n.next
		return 
		
	def isEmpty(self):
		return self.head.next == self.head
	
	def first(self):
		return self.head.next
	
	def last(self):
		return self.head.prev
	
	def deleteNode(self, x):
		if x == None or self.isEmpty():
			return
		x.prev.next = x.next
		x.next.prev = x.prev
	
	def popFront(self):
		if self.isEmpty():
			return
		key = self.head.next.key
		self.deleteNode(self.head.next)
		return key
	
	def popBack(self):
		if self.isEmpty():
			return
		key = self.head.prev.key
		self.deleteNode(self.head.prev)
		return key
	
	def printList(self):
		ch = 'h -> '
		n = self.head.next
		while n != self.head:
			ch += f'{n.key} -> '
			n = n.next
		ch += 'h'
		print(ch)
	
	def findMax(self):
		if self.isEmpty():
			return
		n = self.head.next
		max_v = n.key
		while n != self.head:
			if n.key > max_v:
				max_v = n.key
			n = n.next
		return max_v
	
	def deleteMax(self):
		if self.isEmpty():
			return
		max_v = self.findMax()
		self.deleteNode(self.search(max_v))
		return max_v

	def sort(self):
		new_dl = DoublyLinkedList()
		while self.isEmpty() == False:
			new_dl.pushFront(self.deleteMax())
		return new_dl
```

```python
# 해시 테이블
class HashOpenAddr:
	def __init__(self, size=10):
		self.size = size
		self.keys = [None]*self.size
		self.values = [None]*self.size
	def __str__(self):
		s = ""
		for k in self:
			if k == None:
				t = "{0:5s}|".format("")
			else:
				t = "{0:-5d}|".format(k)
			s = s + t
		return s
	def __iter__(self):
		for i in range(self.size):
			yield self.keys[i]
			
	def find_slot(self, key):
		i = self.hash_function(key)
		start = i
		while self.keys[i] != None and self.keys[i] != key:
			i = (i+1) % self.size
			if i == start:
				return None
		return i
		
	def set(self, key, value=None):
		i = self.find_slot(key)
		if i == None:
			return None
		if self.keys[i] != None:
			self.values[i] = key
		else:
			self.keys[i] = key
			self.values[i] = key
		return key
		
	def hash_function(self, key):
		return key % self.size
	
	def remove(self, key):
		i = self.find_slot(key)
		if self.keys[i] == None:
			return None
		j = i
		while True:
			self.keys[i] = None
			while True:
				j = (j+1) % self.size
				if self.keys[j] == None: return key
				k = self.hash_function(self.keys[j])
				if not((i < k <= j) or (j < i < k) or (k <= j < i)):
					break
			self.keys[i] = self.keys[j]
			i = j
		
	def search(self, key):
		i = self.find_slot(key)
		if self.keys[i] != None:
			return self.values[i]
		else:
			return None
		
	def __getitem__(self, key):
		return self.search(key)
	def __setitem__(self, key, value):
		self.set(key, value)
		
H = HashOpenAddr()
while True:
	cmd = input().split()
	if cmd[0] == 'set':
		key = H.set(int(cmd[1]))
		if key == None: print("* H is full!")
		else: print("+ {0} is set into H".format(cmd[1]))
	elif cmd[0] == 'search':
		key = H.search(int(cmd[1]))
		if key == None: print("* {0} is not found!".format(cmd[1]))
		else: print(" * {0} is found!".format(cmd[1]))
	elif cmd[0] == 'remove':
		key = H.remove(int(cmd[1]))
		if key == None:
			print("- {0} is not found, so nothing happens".format(cmd[1]))
		else:
			print("- {0} is removed".format(cmd[1]))
	elif cmd[0] == 'print':
		print(H)
	elif cmd[0] == 'exit':
		break
	else:
		print("* not allowed command. enter a proper command!")
```

**가장 긴 균형 부문자열 찾기**

```python
def solve(A):
	# code here
	length = 0 # 부문자열 길이가 되어줄 변수
	count_diff = 0 # 0과 1의 개수의 차이
	count_dict = {0: -1} # 0과 1의 개수를 저장할 딕셔너리
	
	for i in range(len(A)):
		if A[i] == '0': # 0이라면 +1
			count_diff += 1
		else: # 1이라면 -1
			count_diff -= 1
			
		if count_diff not in count_dict: # 딕셔너리에 0과 1의 차이가 나온 인덱스를 저장
			count_dict[count_diff] = i
		else:
			# 이미 가지고 있는 index의 차이값과 현재 index의 차이값중 큰 값을 length로 저장
			length = max(length, i - count_dict[count_diff])
	return length
	
A = input().strip()
print(solve(A))

'''
이번 과제에서는 0과 1의 개수를 저장하는데 Dictionary,
그리고 부문자열을 찾는 과정에서 for 반복문을 사용하였다.
알아본 바로는 Dictionary는 O(1)만큼의 수행시간이 걸린다고 한다.
따라서 변수를 저장하는데에 O(1)만큼의 수행시간이 걸렸고,
수업내용을 토대로 for 반복문에서는 O(len(A))만큼의 시간이 걸렸음을 확인할수 있다.
A의 문자열의 개수가 n개라고 한다면
우리가 만든 코드의 총 수행시간은 O(n)으로 결론지을 수 있다.
'''
```
